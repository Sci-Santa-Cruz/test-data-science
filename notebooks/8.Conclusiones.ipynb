## Conclusiones

El análisis permitió identificar patrones consistentes entre **CO₂**, **temperatura**, **humedad** y **hora del día**, reflejando comportamientos asociados tanto a la presencia de personas en la habitación como a los periodos de transición tras su salida. Sin embargo, se observaron discrepancias en algunos días con etiquetas de ocupación, donde variables como **temperatura**, **CO₂** e incluso **luminosidad** mantienen valores similares a los de las muestras etiquetadas como ocupadas.

El modelo de **regresión logística** mostró un desempeño sólido (accuracy 0.92, recall 0.86 para la clase 1), con buena capacidad para distinguir entre espacios ocupados y no ocupados. Las variables más influyentes fueron **CO₂**, **humedad relativa** y **temperatura**, siendo el **CO₂** el predictor dominante.

Los coeficientes mostraron coherencia parcial con el comportamiento físico esperado: el **CO₂** se asoció fuertemente con la ocupación, mientras que la **temperatura** y la **humedad** presentaron signos opuestos a lo esperado, posiblemente por **multicolinealidad**, **efectos no lineales** o **sesgos en la toma de datos**. Aun así, el modelo logra capturar el patrón general de ocupación y concentrar la mayoría de los verdaderos positivos en los primeros deciles del ranking (lift > 5 en el 10 %), confirmando su utilidad para detección o priorización.

El comportamiento del **Cumulative Lift** también indica que el modelo es estable, ya que solo en el último decil su desempeño converge con el comportamiento aleatorio.

En contraste, los experimentos con **K-Means** —más orientados a la exploración que al rendimiento— mostraron que las agrupaciones no capturan adecuadamente los patrones dinámicos asociados a las transiciones de ocupación. Al basarse en proximidad estática, los centroides no reflejan los cambios temporales en temperatura y CO₂. Esto sugiere que, para este tipo de series sensoriales, serían más adecuados métodos que consideren la **dimensión temporal o contextual**, como modelos basados en densidad o secuencias.

**Siguientes pasos:**

* Analizar el comportamiento de los días en los que solo existe una clase.
* Revisar la **representatividad temporal de los datos** (horarios, ventilación, control de clima).
* Incorporar **features cíclicos** que capturen los patrones de transición.
* Aplicar **regularización L2** o **Elastic Net** para estabilizar los coeficientes y reducir correlaciones.
