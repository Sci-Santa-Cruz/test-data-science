# Fase 1: Comprensión del Negocio


### Descripción del Problema de Negocio

**Contexto a alto nivel**
El contexto se basa en un conjunto de datos provenientes de **sensores instalados en una habitación**, los cuales registran variables ambientales como **luz (lux)**, **temperatura (°C)**, **humedad relativa (%)**, **dióxido de carbono (ppm)** y **cociente de humedad**.
El objetivo es **inferir el estado de ocupación de la habitación** (0 = desocupada, 1 = ocupada) a partir de las mediciones ambientales, identificando patrones que indiquen la presencia o ausencia de personas.

---

### Objetivos de Negocio

En esta prueba **no se definieron objetivos de negocio formales** ni métricas de desempeño (KPIs) asociadas.
El propósito principal es **evaluar la capacidad analítica y técnica del candidato** frente a un conjunto de datos sensoriales, abarcando desde la exploración hasta el modelado supervisado y no supervisado.

Cualquier objetivo o métrica adicional planteado en el análisis responde únicamente a un **ejercicio interpretativo**, con el fin de contextualizar la aplicación potencial del modelo (por ejemplo, optimización energética o monitoreo de ocupación).

### Objetivos de Tecnicos

* Identificar las variables que permiten **distinguir condiciones de ocupación** en una habitación.
* Analizar el comportamiento conjunto de las mediciones ambientales y su relación con la **presencia humana**.
* Evaluar la **viabilidad de modelos predictivos**.

---

### Requisitos de Datos Iniciales


**Diccionario de datos (columnas entregadas):**

* `id`: llave primaria.
* `id2`: repetición de `id`.
* `v1`: luz medida en lux.
* `v2`: temperatura en grados Celsius.
* `v3`: fecha/timestamp de medición (`yyyy-MM-dd HH:mm:ss`).
* `v4`: CO₂ en ppm.
* `v5`: humedad relativa en %.
* `v6`: cociente de humedad (kg agua-vapor / kg aire).
* `c1`: variable objetivo binaria — `0` = cuarto NO ocupado, `1` = cuarto ocupado.
